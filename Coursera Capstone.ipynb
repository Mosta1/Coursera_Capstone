```python
# Question 1
import pandas as pd
import numpy as np
import urllib.request
from bs4 import BeautifulSoup                                                            # First we import the necessary libraries
data = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
table = urllib.request.urlopen(data)
soup = BeautifulSoup(table, "lxml")           # then we assign the URL to data, then we parse it and store it in a new variable in Beautiful Soup format
all_tables=soup.find_all("table")             # we use find all to import the table we want into the variable all_tables
all_tables                                     
the_table=soup.find("table", class_="wikitable sortable")
the_table
A=[]
B=[]
C=[]                                        # we add the table data to the_table, then we create 3 empty lists for the 3 columns we want
for row in the_table.findAll("tr"):
    cells=row.findAll("td")
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))             # we use the for loop to import the data from the_table to the empty lists we created
df=pd.DataFrame(A,columns=["PostalCode"])
df["Borough"]=B
df["Neighborhood"]=C                                    # then we create our dataframe and assign the 3 columns names we want
df_filtered = df[df["Borough"] != "Not assigned\n"]    # we then filter the data to remove the rows in which Borough value is Not assigned
canada = df_filtered.sort_values("PostalCode")              # we sort the df postal code column alphabetically and assign it to a new variable
canada                                                      # we finally print the final dataframe
```

```python
# Question 2
Location = pd.read_csv("http://cocl.us/Geospatial_data")       # we import the location data to a dataframe
L = Location.drop(["Postal Code"], axis = 1)                   # we then remove the postal code column and assign the new df to the variable L
Latitude = L["Latitude"]                                       
Longitude = L["Longitude"]                                     # we then import the data of both columns to two new variables
canada["Latitude"] = np.array(Latitude)
canada["Longitude"] = np.array(Longitude)                      # finally we assign these lists after turning them into arrays to our old dataframe
canada                                                         # from question 1
```
